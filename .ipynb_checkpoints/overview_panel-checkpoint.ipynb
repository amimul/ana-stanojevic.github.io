```python
import pandas as pd
import numpy as np
```

```python
folder = "data/expeditions/expedition_"
dfs=[]
for i in range(30):
    dfs.append(pd.read_csv(folder+str(i)+'.csv', sep=';'))
```

```python
map_ocean={'0': 'atlantic', #done
          '1': 'atlantic', #done
          '2': 'atlantic', #done
          '3': 'pacific', #done
          '4': 'pacific', #done
          '5': 'pacific', #done
          '6': 'indian', #done
          '7': 'mediterranean', #done
          '8': 'indian', #done
          '9': 'pacific', #done
          '10': 'atlantic', #done
          '11': 'indian', #done
          '12': 'indian', #done
          '13': 'indian', #done
          '14': 'indian', #done
          '15': 'indian', #done
          '16': 'indian', #done
          '17': 'mediterranean', #done
          '18': 'atlantic', #done
          '19': 'indian', #done
          '20': 'indian', #done
          '21': 'atlantic', #done
          '22': 'pacific', #done
          '23': 'pacific', #done
          '24': 'pacific', #done
          '25': 'pacific', #done
          '26': 'pacific', #done
          '27': 'pacific', #done
          '28': 'pacific', #done
          '29': 'pacific' #done 
          }
```

```python
big_dfs=pd.DataFrame()
for i in range(len(dfs)):
    dfs[i]['ocean'] = map_ocean[str(i)]
    if i==17: dfs[i]=dfs[i].replace('0', np.nan).dropna(subset=['cd1','cd2','cd3','cd4','wd1','wd2','wd3','wd4'], how='all')
    big_dfs = big_dfs.append(dfs[i])  
big_dfs.reset_index(inplace=True, drop=True)
```

```python
big_dfs['wd1'] = big_dfs['wd1'].apply(lambda d: float(str(d).replace(' ', '').replace(',',".")))
big_dfs['wd2'] = big_dfs['wd2'].apply(lambda d: float(str(d).replace(' ', '').replace(',',".")))
big_dfs['wd3'] = big_dfs['wd3'].apply(lambda d: float(str(d).replace(' ', '').replace(',',".")))
big_dfs['wd4'] = big_dfs['wd4'].apply(lambda d: float(str(d).replace(' ', '').replace(',',".")))
big_dfs['cd1'] = big_dfs['cd1'].apply(lambda d: float(str(d).replace(' ', '').replace(',',".")))
big_dfs['cd2'] = big_dfs['cd2'].apply(lambda d: float(str(d).replace(' ', '').replace(',',".")))
big_dfs['cd3'] = big_dfs['cd3'].apply(lambda d: float(str(d).replace(' ', '').replace(',',".")))
big_dfs['cd4'] = big_dfs['cd4'].apply(lambda d: float(str(d).replace(' ', '').replace(',',".")))
```

```python
big_dfs
```

```python
big_dfs = big_dfs.fillna(0)
grouped = big_dfs.groupby(['ocean'])
final = pd.DataFrame(grouped.mean())
final['max_w'] = final[['wd1', 'wd2', 'wd3', 'wd4']].max().max()
final['max_c'] = final[['cd1', 'cd2', 'cd3', 'cd4']].max().max()
```

```python
final[['cd1_r', 'cd2_r', 'cd3_r', 'cd4_r']] = np.sqrt(final[['cd1', 'cd2', 'cd3', 'cd4']]/final['max_c']['atlantic'])
final[['wd1_r', 'wd2_r', 'wd3_r', 'wd4_r']] = np.sqrt(final[['wd1', 'wd2', 'wd3', 'wd4']]/final['max_w']['atlantic'])
final.to_csv('ocean_overview.csv')
```
